```python
df_dsa_p5 = df_dsa.groupby('Segment')['Sales'].sum().reset_index().sort_values(by = 'Sales',
                                                                                      ascending = False)

df_dsa_p5.head()
```

```python
def autopct_format(values): 
    def my_format(pct): 
        total = sum(values) 
        val = int(round(pct * total / 100.0))
        return ' $ {v:d}'.format(v = val)
    return my_format
```

```python
plt.pie(df_dsa_p5['Sales'], 
        labels = df_dsa_p5['Segment'],
        autopct = autopct_format(df_dsa_p5['Sales']),
        startangle = 90)

# Limpa o cÃ­rculo central
centre_circle = plt.Circle((0, 0), 0.82, fc = 'white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Labels e anotaÃ§Ãµes
plt.annotate(text = 'Total de Vendas: ' + '$ ' + str(int(sum(df_dsa_p5['Sales']))), xy = (-0.25, 0))
plt.title('Total de Vendas Por Segmento')
plt.show()
```

```python
df_dsa_p10 = df_dsa.groupby(['Category',
                             'Sub-Category']).sum(numeric_only = True).sort_values('Sales',
                                                                                   ascending = False).head(12)
```

```python
df_dsa_p10 = df_dsa_p10[['Sales']].astype(int).sort_values(by = 'Category').reset_index()
```

```python
df_dsa_p10
```

```python
df_dsa_p10_cat = df_dsa_p10.groupby('Category').sum(numeric_only = True).reset_index()
df_dsa_p10_cat
```

```python
cores_categorias = ['#5d00de',
                    '#0ee84f',
                    '#e80e27']
```

```python
cores_subcategorias = ['#aa8cd4',
                       '#aa8cd5',
                       '#aa8cd6',
                       '#aa8cd7',
                       '#26c957',
                       '#26c958',
                       '#26c959',
                       '#26c960',
                       '#e65e65',
                       '#e65e66',
                       '#e65e67',
                       '#e65e68']
```

```python
fig, ax = plt.subplots(figsize = (18,12))


p1 = ax.pie(df_dsa_p10_cat['Sales'], 
            radius = 1,
            labels = df_dsa_p10_cat['Category'],
            wedgeprops = dict(edgecolor = 'white'),
            colors = cores_categorias)

p2 = ax.pie(df_dsa_p10['Sales'],
            radius = 0.9,
            labels = df_dsa_p10['Sub-Category'],
            autopct = autopct_format(df_dsa_p10['Sales']),
            colors = cores_subcategorias, 
            labeldistance = 0.7,
            wedgeprops = dict(edgecolor = 'white'), 
            pctdistance = 0.53,
            rotatelabels = True)

centre_circle = plt.Circle((0, 0), 0.6, fc = 'white')

fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.annotate(text = 'Total de Vendas: ' + '$ ' + str(int(sum(df_dsa_p10['Sales']))), xy = (-0.2, 0))
plt.title('Total de Vendas Por Categoria e Top 12 SubCategorias')
plt.show()
```
